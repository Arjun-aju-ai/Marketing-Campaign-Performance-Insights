# 3) Exploratory Data Analysis (EDA) and Visualization 

### ➢ Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Acquisition_Cost"], df["ROI"],color="darkorange", alpha=0.5, edgecolors="black")
plt.title("Relationship between Acquisition_Cost and ROI", color="violet", fontsize=20)
plt.xlabel("Acquisition_Cost", color="red", fontsize=16)
plt.ylabel("ROI", color="red", fontsize=16)
plt.grid(True)
plt.show()
```

### ➢ Create a bar chart to visualize the average Conversion_Rate for different
###    Channel_Used, categorized by Campaign_Type.

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
sns.barplot(data=df, x="Channel_Used", y="Conversion_Rate", hue="Campaign_Type", estimator="mean", palette="pastel")
plt.title("average Conversion_Rate for differentChannel_Used, categorized by Campaign_Type")
plt.xlabel("Channel_Used")
plt.ylabel("Conversion_Rate")
plt.show()
```

### ➢ Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Engagement_Score",y="Campaign_Type",data=df,color="green")
plt.title("Distribution of Engagement_Score across different Campaign_Type", color="red",fontsize=20)
plt.xlabel("Engagement_Score",color="blue",fontsize=12)
plt.ylabel("Campaign_Type",color="blue",fontsize=12)
plt.tight_layout()
plt.show()
```

### ➢ Analyze the average ROI by Company using a bar chart to compare the profitability
###        of campaigns conducted by different companies.

```python
fig,ax=plt.subplot(figsize=(10,8))
fig.set_facecolor("light blue")
ax.set_facecolor("light green")
plt.title("Average ROI by Company")
plt.xlabel("Company")
plt.ylabe("Average ROI")
sns.barplot(data=df,x="Company",y="ROI",hue="Company",estimator="mean",palette="pastel")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
fig.set_facecolor('lightgrey')
ax.set_facecolor('purple')
plt.title("Average ROI by Company")
plt.xlabel("Company")
plt.ylabel("Average ROI")
plt.xticks(rotation=45)  
sns.barplot(data=df,x="Company",y="ROI",hue="Company",estimator="mean",palette="pastel")
plt.show()
```

### ➢ Examine the correlation between Engagement_Score and Conversion_Rate
### using a heatmap.

```python
cr=df[["Engagement_Score","Conversion_Rate"]].corr()
plt.figure(figsize=(10,6))
sns.heatmap(cr,annot=True,cmap="BuGn",linewidths=0.8)
plt.title("correlation:  between Engagement_Score and Conversion_Rate",color="Blue",fontsize=20,pad=20)
plt.xlabel('Metrics', color='red', fontsize=16, labelpad=10)
plt.ylabel('Metrics', color='red', fontsize=16, labelpad=10)
plt.tight_layout()
plt.show()
```

### Language using a bar chart.

```python
group= df.groupby(['Language', 'Customer_Segment'])['Conversion_Rate'].mean().reset_index()
h_segment=group.loc[group.groupby('Language')['Conversion_Rate'].idxmax()]
sns.barplot(data=h_segment,x="Language",y="Conversion_Rate",hue="Customer_Segment",palette="pastel")
plt.title("Highest Conversion_Rate",color="cyan",fontsize=20)
plt.xlabel("Language",color="yellow",fontsize=16)
plt.ylabel("highest Conversion_Rate for each Language",color="yellow",fontsize=16)
plt.legend(title="Customer_Segment")
plt.show()
```

# Time-BaseAnalysis


### ➢ Plot the distribution of Duration using a histogram.

```python
plt.hist(df["Duration"], bins=12, color='orange', edgecolor='green')
plt.show()
```

### ➢ Analyze how the overall Conversion_Rate has changed over Date for each
### Company using a line chart.

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
conversion_trend = df.groupby(["Date", "Company"])["Conversion_Rate"].mean().reset_index()
sns.lineplot(data=conversion_trend, x="Date", y="Conversion_Rate", hue="Company", marker='o')
plt.title("Conversion Rate Trend Over Time by Company")
plt.xlabel("Date")
plt.ylabel("Average Conversion Rate")
plt.show()
```

### ➢ Examine the trend of Engagement_Score over Date with a line chart.

```python


df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])
df = df.sort_values(by='Date')  

df_daily = df.groupby('Date')['Engagement_Score'].mean().reset_index()

plt.figure(figsize=(12, 6))
#plt.plot(df_daily['Date'], df_daily['Engagement_Score'], color='teal')
sns.lineplot(data=df_daily, x='Date', y='Engagement_Score', marker='o', color='blue')
plt.title('Engagement Score Trend Over Time', color='red', fontsize=16, pad=20)
plt.xlabel('Date', color='orange', fontsize=13, labelpad=10)
plt.ylabel('Average Engagement Score', color='orange', fontsize=13, labelpad=10)

plt.tight_layout()
plt.grid(True)
plt.show()
```

# Geographic Analysis


### ➢ Determine which location has the highest Acquisition_Cost using a bar chart.

```python
location_cost = df.groupby("Location")["Acquisition_Cost"].mean().reset_index()
location_cost = location_cost.sort_values(by="Acquisition_Cost", ascending=False)
sns.barplot(data=location_cost, x="Location", y="Acquisition_Cost", palette="magma")
plt.title("Average Acquisition Cost by Location")
plt.xlabel("Location")
plt.ylabel("Average Acquisition Cost")
plt.xticks(rotation=45)
plt.show()
```

### ➢ Visualize the Conversion_Rate by different Location, categorized by
### Target_Audience, using a bar chart.

```python
location_conversion = df.groupby(["Location", "Target_Audience"])["Conversion_Rate"].mean().reset_index()
sns.barplot(data=location_conversion, x="Location", y="Conversion_Rate", hue="Target_Audience", palette="pastel")
plt.title("Average Conversion Rate by Location and Target Audience")
plt.xlabel("Location")
plt.ylabel("Average Conversion Rate")
plt.legend(title="Target Audience")
plt.show()
```

### ➢ Illustrate the proportion of ROI by Location using a pie chart.

```python
location_roi = df.groupby("Location")["ROI"].sum().reset_index()
labels = location_roi["Location"]
sizes = location_roi["ROI"]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title("Proportion of Total ROI by Location")
plt.axis('equal') 
plt.show()
```
